# Cruce de información

Usando las coordenadas de los centros penitenciarios se cruzó la información de MinJusticia con la del INPEC.

También tenemos la matriz de distancias euclideanas entre los centros penitenciarios, y el cruce de este con la base de datos del final está comentado.

```python
poblacion_pospenada_df = pd.read_csv('Data/Poblaci_n_pospenada_que_ha_sido_reincidente.csv')
centros_penitenciarios_df = pd.read_csv('Data/centros_penitenciarios_dptos.csv')
shdi_df = pd.read_excel('Data/SHDI.xlsx')
distances_df = pd.read_csv('Data/Matriz_distancias.csv')

#Como las proyecciones entre los polígonos del INPEC y nuestro polígono (https://sites.google.com/site/seriescol/shapes)
#no era exactas, cuadramos los dos NA que tenemos
centros_penitenciarios_df.loc[centros_penitenciarios_df.name == '101 EPMSC Leticia', 'NOMBRE_DPT'] = 'amazonas'
centros_penitenciarios_df.loc[centros_penitenciarios_df.name == '401 EPMSC Arauca', 'NOMBRE_DPT'] = 'arauca'

#Normalizamos y eliminamos los departamentos que tienen información entre paréntesis
shdi_df.region = shdi_df.region.apply(lambda x: re.sub("[\(].*?[\)]", "", str(x)).lower().strip())
centros_penitenciarios_df.NOMBRE_DPT = centros_penitenciarios_df.NOMBRE_DPT.apply(lambda x: str(x).lower().strip())

#Creamos un diccionario que nos permita cruzar entre departamentos
arr_sentence = {}
i = 0
max_distance = 161616
for sentence_i in centros_penitenciarios_df.NOMBRE_DPT.unique():
    last_dist = max_distance
    for sentence_j in shdi_df.region.unique():
        dist = levenshtein(sentence_i, sentence_j)
        if dist < last_dist:
            final_sentence = [sentence_i, sentence_j]
            last_dist = dist
    arr_sentence[final_sentence[0]] = final_sentence[1]
    i+=1

#print(arr_sentence)
#Después de una breve inspección visual del diccionario encontramos que:
#Nada que hacer, San Andrés tiene un nombre demasiado largo
arr_sentence['archipielago de san andres providencia y santa catalina'] = 'san andres'
#Valle del Cauca podría ser Valle o ser Cauca
arr_sentence['valle del cauca'] = 'valle'

#Creamos la llave en el dataframe de centros penitenciarios y cruzamos
centros_penitenciarios_df['region'] =  centros_penitenciarios_df.NOMBRE_DPT.apply(lambda x: arr_sentence[x])
shdi_df = shdi_df.merge(centros_penitenciarios_df, on = 'region', how = 'left')

#Normalizamos y eliminamos los números de los centros penitenciarios
shdi_df['key_i'] = shdi_df.name.apply(lambda x: re.sub("\d+", "", str(x)).lower().strip())
poblacion_pospenada_df['key_j'] = poblacion_pospenada_df.ESTABLECIMIENTO.apply(lambda x: 
                                        re.sub("[\(].*?[\)]", "", str(x)).lower().strip())

shdi_df['key_i'] = shdi_df.key_i.apply(lambda x: re.sub("-", " ", str(x)).strip())
poblacion_pospenada_df['key_j'] = poblacion_pospenada_df.key_j.apply(lambda x: re.sub("-", " ", str(x)).strip())

#Eliminamos algunas palabras que dificultan crear una llave entre dataframes
stop = set(['epc', 'cpms', 'cpams', 'epmsc-rm', 'ere', 'cmps', 'no.', 'rm', 'ep', 'epms', 'ec', 'epmsc', 
 'rm', 'de', 'la', 'el', 'y', 'epams', 'epamscas', 'complejo', 'carcelario', 'penitenciario', 'metropolitano',
'cms', 'cocuc', 'condenados', 'sindicados', 'mujeres', 'coiba', 'r.', 'pofac', 'mediana', 'seguridad', 'cojam'])
shdi_df['key_i'] = shdi_df.key_i.apply(lambda x: ' '.join([w for w in x.split() if w not in stop]))
poblacion_pospenada_df['key_j'] = poblacion_pospenada_df.key_j.apply(lambda x: ' '.join([w for w in x.split() if w not in stop]))

#Creamos un diccionario que nos permita cruzar entre departamentos
arr_sentence = {}
i = 0
max_distance = 161616
for sentence_i in poblacion_pospenada_df.key_j.unique():
    last_dist = max_distance
    for sentence_j in shdi_df.key_i.unique():
        dist = levenshtein(sentence_i, sentence_j)
        if dist < last_dist:
            final_sentence = [sentence_i, sentence_j]
            last_dist = dist
    arr_sentence[final_sentence[0]] = final_sentence[1]
    i+=1

#print(arr_sentence)
#Después de una breve inspección visual del diccionario encontramos que:
arr_sentence['esperanza guaduas'] = 'guaduas esperanza'
arr_sentence['picaleña'] = 'ibague'
arr_sentence['ibague picaleña'] = 'ibague'
arr_sentence['manzanares'] = 'pensilvania'
arr_sentence['cienaga'] = 'santa marta'
arr_sentence['zapatoca'] = 'bucaramanga'

#Creamos la llave en el dataframe de poblacion pospenada
poblacion_pospenada_df['key_i'] = poblacion_pospenada_df.key_j.apply(lambda x: arr_sentence[x])
#Eliminamos casos en los que puede haber varias cárceles cerca, como en medellín, bogotá o ibagué,
#después de todo la información regional es la misma para todas
shdi_without_duplicates_df = shdi_df.drop_duplicates(['year', 'key_i'])
#Cruzamos ambos dataframes
poblacion_pospenada_df = poblacion_pospenada_df.merge(shdi_without_duplicates_df, left_on = ['AÑO_INGRESO', 'key_i'], 
                                                      right_on = ['year', 'key_i'], how = 'left')

#Cruzamos población pospenada con la matriz de distancias
#poblacion_pospenada_df = poblacion_pospenada_df.merge(distances_df, left_on = ['name'], right_on = ['ID'], how = 'left')
```

# Análisis de los crímenes

Para estudiar cómo se relacionan los crímenes entre sí 

- se calcularon la cantidad de veces que alguien cometió un delito

- luego se calculó la matriz de correlación entre estos crímenes

- definimos una distancia, al normalizar la correlación para que estuviera acotada inferiormente, como $$d_{ij} = \max(cor_{ij}) - cor_{ij}$$

- con ésta, definimos una medida de similaridad dada por $$s_{ij} = 1 - \frac{d_{ij}}{\max(d_{ij})}$$ 

- decimos que dos crímenes $\left[c_{i}, \enspace c_{j}\right]$ (dos nodos) están relacionados cuando $$d_{ij} < \tau$$ donde $\tau$ es un umbral arbitrario

- el umbral óptimo $\tau^*$ es aquel que hace que el grafo sea libre de escala; esto es, $$p(k) = k^{-\lambda}, \text{ donde $\lambda$ está entre 3 y 2, y $k$ es la proporción de nodos de grado $k$}$$

- a este grafo de delitos $$G = \{[c_i], \enspace 1[d_{ij} < \tau^*]\}$$ podemos hacerle análisis de comunidades y ver cuáles son los delitos más relacionados entre sí

```python
crime = poblacion_pospenada_df[['INTERNO', 'DELITO']]
crime = crime.groupby(['INTERNO', 'DELITO']).size().reset_index()
crime.columns = ['id', 'crime', 'count']
crime = crime.pivot(index='id', columns='crime', values = 'count').fillna(0)
corr = crime.corr()
```

```python
edge_dict = {}
h = 0
for i in np.arange(corr.shape[0]):
    for  j in np.arange(corr.shape[1]):
        if i > j:
            edge_dict[h] = [corr.columns[i], corr.columns[j], corr.iloc[i, j]]
        h+=1

pd_edges = pd.DataFrame.from_dict(edge_dict).T
pd_edges.columns = ['source', 'target', 'weight']
pd_edges.weight = pd_edges.weight.max() - pd_edges.weight
```

```python
sns.distplot(pd_edges.weight)
```

```python
grid = np.linspace(0.6, 0.7, num = 161)

def free_scale(t):
    pd_edges['reduced_weight'] = pd_edges.weight
    pd_edges.loc[pd_edges.weight > t, 'reduced_weight'] = 0
    pd_edges['similarity'] = 1 - pd_edges.reduced_weight/pd_edges.reduced_weight.max()
    pd_edges.loc[pd_edges.similarity == 1, 'similarity'] = 0
    graph = nx.from_pandas_edgelist(pd_edges.loc[pd_edges.similarity > 0, ['source', 'target', 'weight']])
    degree_sequence = sorted([b for a, b in graph.degree()], reverse = True)
    fit = plw.Fit(degree_sequence)
    return(fit.power_law.alpha)

result = map(free_scale, grid)

alpha = pd.Series(result)
```

```python
alpha.plot()
```

```python
distance_threshold = grid[56]

pd_edges['reduced_weight'] = pd_edges.weight
pd_edges.loc[pd_edges.weight > distance_threshold, 'reduced_weight'] = 0

pd_edges['similarity'] = 1 - pd_edges.reduced_weight/pd_edges.reduced_weight.max()
pd_edges.loc[pd_edges.similarity == 1, 'similarity'] = 0

graph = nx.from_pandas_edgelist(pd_edges.loc[pd_edges.similarity > 0, ['source', 'target', 'weight']])
degree_sequence = sorted([b for a, b in graph.degree()], reverse = True)
fit = plw.Fit(degree_sequence)

print(fit.power_law.alpha)

sns.distplot(degree_sequence)
```

```python
edges_result = pd_edges.loc[pd_edges.similarity > 0, ['source', 'target', 'similarity']]

partition = community_louvain.best_partition(graph)

nodes_result = pd.DataFrame.from_dict(partition.items())
nodes_result.columns = ['Id', 'community'] 
nodes_result['label'] = nodes_result['Id']
```

```python
# Estos son los archivos con los que se construye el grafo en Gephi
edges_result.to_csv('Data/edges_result.csv', index = False)
nodes_result.to_csv('Data/nodes_result.csv', index = False)
```

# Análisis de los datos con información demográfica

```python
datos = pd.read_excel('Data/ReincidenciaPospenadosNal2010-31Mayo2020Rev_con variables demograficas.xlsx', header = 6)
```

```python
from datetime import datetime

def convertir_fecha(x):
    if str(x) == 'nan':
        res = np.datetime64('NaT')
    else:
        res = datetime.strptime(x[0:6]+'20'+x[6:8], "%d/%m/%Y")
    
    return res

for column in ['FECHA_INGRESO', 'FECHA_SALIDA', 'FECHA_CAPTURA']:
    datos[column] = datos[column].apply(convertir_fecha)
    
    datos = datos.sort_values(['INTERNOEN', column], ascending = False)
    
    datos['DIAS' + column[5:]] = -1*datos[column].diff().copy()

    datos.loc[(datos.INTERNOEN != datos.INTERNOEN.shift(1)) | (datos['DIAS' + column[5:]] == np.timedelta64(0)), 
              ['DIAS' + column[5:]]] = datetime.today() - datos[column]
    
datos['DIAS_CONDENA'] = datos['FECHA_SALIDA'] - datos['FECHA_INGRESO'] 
datos['DIAS_JUDICIALIZACION'] = datos['FECHA_INGRESO'] - datos['FECHA_CAPTURA']
datos['DIAS_LIBRE'] = datos['DIAS_INGRESO'] - datos['DIAS_CONDENA']
```
